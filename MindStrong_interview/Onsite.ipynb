```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

% matplotlib inline
```

```python
sample1 = np.random.normal(loc = -0.69, scale=0.3, size=200)
sample2 = np.random.normal(loc = -0.3, scale=0.3, size=200)
```

```python
sns.distplot(np.exp(sample1), bins=30, rug=True, label='Before')
sns.distplot(np.exp(sample2), bins=30, rug=True, label='After')
plt.legend()
plt.xlabel('Syanptic Event Amplitude (nA)')
plt.ylabel('Counts')
plt.xlim([0, 1.8])
```
