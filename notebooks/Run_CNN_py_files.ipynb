Run 3 versions of training data set with LeNet5

Data input and network structure is as follows:

Trainin data: 
data: [Sample_size, imgsize * imgsize]
label: [Sample_szie, classsize]

Input: 28 * 28 * 1
Cov layer 1: 5 * 5 * 32 filters maps followed by Relu
output_1: 25 * 25 * 32
Max pooling: sub sampling by 2
output_1_maxpool: 13 * 13 * 32
Conv layer 2: 5 * 5 * 64 filter maps followed by Relu
output_2: 11 * 11 * 64
Max pooling subampling by 2
output_2_max: 6 * 6 * 64
Fully connected layer (add one bias neuron??): 7 * 7 * 64
output: 1024
Read out layer: output 10
